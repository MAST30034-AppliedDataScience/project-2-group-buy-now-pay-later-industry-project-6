```python
# Function to identify non-integer values THIS IS FROM CHATGPT
def find_non_integer_rows(df, column_name):
    # Attempt to convert to numeric, coercing errors to NaN
    df[column_name + '_int'] = pd.to_numeric(df[column_name], errors='coerce')
    
    # Identify rows where conversion failed
    non_integer_rows = df[df[column_name + '_int'].isna()]
    
    # Optionally drop the helper column
    df.drop(columns=[column_name + '_int'], inplace=True)
    
    return non_integer_rows

for col in ['sum', 'median', 'mean']:
    non_integers = find_non_integer_rows(df, col)
    if not non_integers.empty:
        print(f"Potential non-integer values in column '{col}':")
        print(non_integers[[col]])  # Display the problematic rows
        print()
```

```python
# Need to change np rows to integers. Will replace with the mean value

# Function to clean and convert column values to integers
def clean_and_convert_to_int(df, column_name):
    # Remove commas or other non-numeric characters if necessary
    df[column_name] = df[column_name].astype(str).str.replace(',', '')  # Example for removing commas

    # Convert to numeric, coercing errors to NaN
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce')

    # Convert to integer, replacing NaNs with a default value (e.g., 0)
    df[column_name] = df[column_name].fillna(df[column_name].mean()).astype(int)

for col in ['sum', 'median', 'mean']:
    clean_and_convert_to_int(df, col)
```
