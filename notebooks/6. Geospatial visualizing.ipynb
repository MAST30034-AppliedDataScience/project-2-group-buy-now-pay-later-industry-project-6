# Geospatial Visualization
In this notebook we complete a geospatial visualization of Australia's SA2 districts with our areas of interest.
Due to the large size of the maps created, they have been commented out of the code.
Feel free to uncomment to view the maps! Screenshots have been included.

```python
import os
import sys
import pandas as pd
sys.path.append("../")
from pyspark.sql import SparkSession
from pyspark.sql.functions import avg, count, sum
from pyspark.sql.functions import col
import geopandas as gpd
import folium

# Start spark session
spark = (
    SparkSession.builder.appName("geospatial")
    .config("spark.sql.repl.eagerEval.enabled", True) 
    .config("spark.sql.parquet.cacheMetadata", "true")
    .config("spark.sql.session.timeZone", "Etc/UTC")
    .getOrCreate()
)

# Read in transactions data
transactions = spark.read.parquet("../data/tables/synthetic/transactions")

# Read in consumer data (from notebook 3)
consumers = spark.read.csv("../data/curated/consumers_median_income", header=True, inferSchema=True)

# Read in consumer_id data
consumer_id_info = spark.read.parquet("../data/tables/synthetic/consumer_user_details.parquet")
```

```python
# Read in shapefile for Australia SA2 districts
sf = gpd.read_file("../data/tables/abs/SA2_2021_AUST_GDA2020.shp")

# Convert the geometry shape to latitude and longitude
sf['geometry'] = sf['geometry'].to_crs("+proj=longlat +ellps=WGS84 +datum=WGS84 +no_defs")

# Create JSON
geoJSON = sf[['SA2_CODE21','geometry']].drop_duplicates('SA2_CODE21').to_json()
```

```python
# Read in ABS income data and clean it (taken from notebook 3)
abs_df = pd.read_csv("../data/tables/abs/ABS_2021.csv")

# We only take SA2 data and data from 2021
abs_filtered = abs_df[(abs_df["REGION_TYPE: Region Type"] == "SA2: Statistical Area Level 2") & 
                      (abs_df["TIME_PERIOD: Time Period"] == 2021)]

abs_filtered["REGION: Region"] = abs_filtered["REGION: Region"].str[:9].astype("int64")

abs_filtered = abs_filtered.drop(["DATAFLOW", "REGION_TYPE: Region Type", "STATE: State",
                                  "TIME_PERIOD: Time Period"], axis=1) # drop useless columns
abs_filtered = abs_filtered.rename(columns={"MEDAVG: Median/Average": "statistic",
                                            "REGION: Region": "region",
                                            "OBS_VALUE": "value"}) # rename columns

# Resetting the index
abs_filtered = abs_filtered.pivot_table(index="region", columns="statistic", values="value", aggfunc="sum")
abs_filtered = abs_filtered.reset_index()

# Rename columns for simplicity
abs_filtered.columns = ["region", "median_age", "median_personal_income",
                        "median_family_income", "median_household_income", "median_mortgage",
                        "median_rent", "avg_bedroom", "avg_household"]

# Remove outliers and zero values
abs_filtered = abs_filtered[(abs_filtered["median_age"] > 0) & 
                            (abs_filtered["median_personal_income"] > 0) &
                            (abs_filtered["median_family_income"] > 0) &
                            (abs_filtered["median_household_income"] > 0) &
                            (abs_filtered["median_mortgage"] > 0) &
                            (abs_filtered["median_rent"] > 0) &
                            (abs_filtered["avg_bedroom"] > 0) &
                            (abs_filtered["avg_household"] > 0)]

abs_filtered = abs_filtered.dropna() # remove NaN values
```

## Choropleth 2: SA2 Districts with Total Number of Orders

```python
# Need the user_id column with our consumer info as user_id is in the transactions dataframe
consumers_joined = consumers.join(consumer_id_info, on="consumer_id", how="left")

# Dropping columns we are not interested in
user_id_region = consumers_joined.drop("consumer_id","name","postcode","address","gender","median_personal_income","state")

# Attaching each transaction with the SA2 district of the consumer
transactions_with_region = transactions.join(user_id_region, on="user_id", how="left")

# Aggregating by region to get the number of orders per SA2 district
region_order_count = transactions_with_region.groupby("region").agg(count("region").alias("order_count"))

# Changing the region column back to integers as it was in scientific notation
region_order_count = region_order_count.withColumn("region", col("region").cast("int"))

# Converting to a pandas dataframe to create the choropleth
region_order_count_pandas = region_order_count.toPandas()
```

## Choropleth 3: SA2 Districts with Total Order Dollar Value

```python
# Aggregating by region to get the total dollar value per SA2 district
region_dollar_val_count = transactions_with_region.groupby("region").agg(sum("dollar_value").alias("total_order_dollar_value"))

# Taking a log transformation of the total dollar value for better visualisation
region_dollar_val_count = region_dollar_val_count.withColumn("log_total_order_dollar_value", log(region_dollar_val_count["total_order_dollar_value"]))

# Converting to a pandas dataframe to create the choropleth
region_dollar_val_count_pandas = region_dollar_val_count.toPandas()
```
